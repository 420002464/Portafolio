# Ley Fuerte de los Grandes Números

Esta ley dice que: $$\lim_{n\to\infty}P\left(|\bar{X_n}-\mu|<\varepsilon\right)=1$$ para cualquier $\varepsilon>0$

Veremos ejemplos para diferentes distribuciones.

## Normal

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
def norm(mu, sig, n):
    X = np.random.normal(loc=mu, scale=sig, size=n)
    return X
Y = [np.mean(norm(0,1,i)) for i in range(1,1500)]
X = [i for i in range(1,1500)]
sns.set_style("darkgrid")
ax = sns.lineplot(x=X,y=Y,color='pink')
sns.lineplot(x=[1,1500],y=0, color='red')
ax.set(xlabel='$n$', ylabel=r'$\overline{X}$')
```

En el gráfico podemos ver que para una muestra aleatorio de $\mathcal{N}(0,1)$ vemos que conforme $n$ es más grande se va acercando al valor teórico de $\mu$, que en este caso es $\mu=0$

## Gamma

```python
def gam(a,l,n):
    return np.random.gamma(a, 1/l, size=n)
alpha = 1
lamb = 2
Y = [np.mean(gam(alpha,lamb,i)) for i in range(2,1500)]
X = [i for i in range(2,1500)]
sns.set_style("darkgrid")
sns.lineplot(x=X,y=Y,color='fuchsia')
sns.lineplot(x=[1,1500], y=alpha/lamb, color='red')
plt.xlabel('$n$')
plt.ylabel(r'$\overline{X}$')
```

En este caso caso tenemos que si $X\sim\Gamma(\alpha,\lambda)$, tenemos que $E[X]=\frac{\alpha}{\lambda}$. Teóricamente como $\alpha=1$ y $\lambda=2$, tenemos que $\mu=\frac{1}{2}$ y en efecto vemos que el promedio se va acercando a la media teórica según aumenta $n$.

## Bernoulli

```python
def ber(p, n):
    return np.random.choice(a=[0,1],p=[1-p,p],size=n)
p=0.6
Y = [np.mean(ber(p, i)) for i in range(2,1500)]
X = [i for i in range(2,1500)]
sns.set_style("darkgrid")
sns.lineplot(x=X,y=Y,color='mediumspringgreen')
sns.lineplot(x=[1,1500], y=p, color='red')
plt.xlabel('$n$')
plt.ylabel(r'$\overline{X}$')
```

En este ejemplo tenemos que $X\sim Ber(p)$, entonces $E[X]=p$. En el código $p=0.6$, entonces $\mu=0.6$, que es a lo que se va acercando la media conforme aumenta $n$.

## Poisson

```python
def poi(λ, n):
    return np.random.poisson(lam=λ, size=n)
λ = 6.75 
Y = [np.mean(poi(λ, i)) for i in range(2,1500)]
X = [i for i in range(2,1500)]
sns.set_style("darkgrid")
sns.lineplot(x=X,y=Y,color='black')
sns.lineplot(x=[1,1500], y=λ, color='red')
plt.xlabel('$n$')
plt.ylabel(r'$\overline{X}$')
```

Al ser una distribución de $Poi(\lambda)$, tenemos que $\mu=\lambda$, entonces conforme $n$ crezca nos vamos a ir acercando a $6.75$.
