# Copo de Nieve de Koch

```python
import matplotlib.pyplot as plt
import numpy as np
```

Una de las funciones que crearemos para simplificar el trabajo es `linea`, donde resumirá la gráficación de `matplotlib`.

```python
def linea(xa, ya, xb, yb):
    plt.plot([xa, xb], [ya, yb])
```

Usando `linea`, programaremos `Curva_von_Koch`, la cual hará que para una línea dada se hagan triángulos equilateros a $\frac{1}{3}$ de distancia de cada vértice existente. En esta nueva función se está usando la recursión.

```python
def Curva_von_Koch(xa, ya, xb, yb, n):
    
    if n==0:
        linea(xa, ya, xb, yb)
    #voy a probar sin else
    else:
        xc = xa + (xb-xa)/3
        yc = ya + (yb-ya)/3
        xd = xb - (xb-xa)/3
        yd = yb - (yb-ya)/3
        
        xe = (xc+xd)*np.cos(np.pi/3) - (yd-yc)*np.sin(np.pi/3)
        ye = (yc+yd)*np.cos(np.pi/3) + (xd-xc)*np.sin(np.pi/3)
        
        Curva_von_Koch(xa, ya, xc, yc, n-1)
        Curva_von_Koch(xc, yc, xe, ye, n-1)
        Curva_von_Koch(xe, ye, xd, yd, n-1)#Con n te salen triangulos interesantes
        Curva_von_Koch(xd, yd, xb, yb, n-1)
        #print(xa,xe)
        #plt.show()#Me da los triangulos uno por uno
    #plt.show() me da cada segmento
```

Usando `Curva_von_Koch` programaremos `Copo_von_Koch`, donde se fijarán aristas en los puntos $(-1,0)$, $(1,0)$ y $(0,\sqrt{3})$, que serán los vértices en donde se va a usar la función `Curva_von_Koch`, lo que resultará en la formación posterior del copo por medio de iteraciones.

```python
def Copo_von_Koch(n):
    x1, y1 = -1, 0
    x2, y2 = 1, 0
    x3, y3 = 0, np.sqrt(3)
    
    Curva_von_Koch(x2, y2, x1, y1, n)
    Curva_von_Koch(x3, y3, x2, y2, n)
    Curva_von_Koch(x1, y1, x3, y3, n)#De x3,y3 a x1,y1 va de los 3's a los 1's y como el orden importa hace que el copo se vea sumido para dentro o invertido
```
