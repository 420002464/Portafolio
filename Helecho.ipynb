# Fractal Helecho

Primero importaremos las bibliotecas `numpy` y `matplotlib` para hacer los cálculos y graficar.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Definiremos la función `helecho` para graficar el fractal. Al ser un fractal generado por una distribución uniforme se usarán `if` y `elif` para clasificar y calcular cada iteración según el resultado obtenido de la distribución uniforme.

```python
def helecho(n) -> np.array:
    #Generación de números
    Xk = np.array([[0.5], [0.5]])
    X = []
    Y = []

    for i in range(n):
        a = np.random.rand()
        if(0<a<.85):
            Xk = np.array([[0.85, 0.04], [-0.04, 0.85]])@Xk + np.array([[0], [1.6]])
            X.append(*Xk[0])
            Y.append(*Xk[1])
        elif(.85<=a<.92):
            Xk = np.array([[0.20, -0.26], [0.23, 0.22]])@Xk + np.array([[0], [1.6]])
            X.append(*Xk[0])
            Y.append(*Xk[1])
        elif(.92<=a<.99):
            Xk = np.array([[-0.15, 0.28], [0.26, 0.24]])@Xk + np.array([[0], [0.44]])
            X.append(*Xk[0])
            Y.append(*Xk[1])
        elif(.99<=a<1):#También probar else
            Xk = np.array([[0, 0], [0, 0.16]])@Xk + np.array([[0], [0]])
            X.append(*Xk[0])
            Y.append(*Xk[1])
    plt.figure(figsize=(6,8))
    plt.scatter(X, Y, s=0.2, color = 'green')
    plt.show()
```

En la función, `n` son los puntos o iteraciones del helecho, entre más puntos haya, el fractal se irá pareciendo más a un helecho.

```python
helecho(30000)
```
