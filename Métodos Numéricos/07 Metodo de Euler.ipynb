# Método de Euler

Este método ayuda a resolver numéricamente Ecuaciones Diferenciales Ordinarias.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns #Se importa matplotlib porque ésta tiene comandos que seaborn no tiene
sns.set_theme()
```

```python
def euler(x0, xf, y0, h=0.5):
    X = np.arange(x0, xf + h, step=h)
    Y = np.array([y0])
    n = len(X)
    for i in range(n-1):
        yi = Y[i] + h*f(X[i], Y[i])
        Y = np.append(Y, yi)
    return X,Y
```
