# Método Runge-Kutta4

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

La EDO para la cual aplicaremos este método es $$y'=x\cdot y$$

```python
def f(x,y):
    return x*y
```

```python
def rk4(x0, xf, y0, h=0.5):
    X = np.arange(x0, xf + h,h)
    Y = np.array([y0])
    n = len(X)
    for i in range(n-1):
        k1 = f(X[i], Y[i])
        k2 = f(X[i] + 0.5*h, Y[i] + 0.5*h*k1)
        k3 = f(X[i] + 0.5*h, Y[i] + 0.5*h*k2)
        k4 = f(X[i] + h, Y[i] + h*k3)
        yi = Y[i] + h/6*(k1 + 2*k2 + 2*k3 + k4)
        Y = np.append(Y, yi)
    return X, Y
```

```python
C = rk4(0, 3, 1)
X=np.linspace(0, 3,100)
sns.lineplot(x=C[0], y=C[1], color="g", label="RK4")
plt.plot(X, 1*np.exp(X**2/2), "r", label="$e^{\\frac{x^2}{2}}$")
plt.legend()
plt.show() 
```

Ahora haremos que el tamaño del paso, i.e. $h$ sea más pequeño.

```python
C = rk4(0, 3, 1, 0.3) #el nuevo h es de 0.3
X=np.linspace(0, 3,100)
sns.lineplot(x=C[0], y=C[1], color="g", label="RK4")
plt.plot(X, 1*np.exp(X**2/2), "r", label="$e^{\\frac{x^2}{2}}$")
plt.legend()
plt.show() 
```
