# Normal multivariada

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Debido a que se limita a 2 coordenadas, se simplifica la función de distribución de la normal multivariada de 2 dimensiones.

Donde se tomó $\rho=0$, $\sigma_x=1$ y $\sigma_y=1$, es decir, son variables independiente la una de la otra. Por ende la función de distribución resulta en: 
$$f(x_1,x_2)=\frac{1}{2\pi}e^{-\frac{1}{2}\left(x_1^2+x_2^2\right)}$$
Recordemos que también podemos ver a $x_1$ y $x_2$ como $x$ y $y$ respectivamente

```python
def NME(x1, x2):
    return 1/(2*np.pi)*np.exp(-1/2 * (x1**2 + x2**2))

x1 = x2 = np.linspace(-2.5, 2.5, 100)

X1, X2 = np.meshgrid(x1, x2)

Z = NME(X1, X2)
print(Z)

fig = plt.figure()
# En algunas versiones hay que poner auto_add_to_figure=False
ax = Axes3D(fig)  # ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)

ax.plot_surface(X1, X2, Z, cmap='rainbow', edgecolor='black')#gist_stern  rainbow   cool con edgecolor="black"

plt.show()
```
